## RPy: 通过 Python 调用 R 的功能
- `Python` 作为胶水语言，可以通过 `RPy` 来调用 `R` 的功能
- 或者，可以使用 Shell 来整合 `Python` 和 `R` 的程序
  - 使用 `Makefile` 将不同语言的输入和输出连接
- 例子：连接各种各样的语言和工具
    - http://lgautier.github.io/odsc-pdaibp-slides/

## 集合运算
- 交、并、差，笛卡尔积

```python
from rpy2.robjects import r
A = r('A <- data.frame(ID=c(1,2), name=c("Wang", "Li"))')
A
```

```python
B = r('B <- data.frame(ID=c(2,3), name=c("Li", "Zhang"))')
B
```

```python
r.merge(A,B)
```

```python
print(r.rbind(A,B))
```

```python
from rpy2.robjects.lib.dplyr import setdiff
setdiff(A,B)
```

```python
AxB = r.merge(A,B,by=r("NULL"))
AxB
```

## 关系运算
- 投影，选择，连接，除法

```python
Am = r.cbind(A, GPA=r("c(3,4)"))
Am
```

```python
Bm = r.cbind(B, age=r('c(19,20)'))
Bm
```

```python
r.merge(Am,Bm,all=r("TRUE"))
```

```python
r.merge(Am,Bm,all=r("TRUE"))
# somehow "all_x" is not correct, 左连接不可用
```

## 关系代数之外的常用运算
- GroupBy, SortBy

```python
from rpy2.robjects.lib.dplyr import DataFrame, group_by, summarise

dAxB = DataFrame(AxB)
dAxB
```

```python
(dAxB.group_by("name.x").summarise(n='n()'))
```
